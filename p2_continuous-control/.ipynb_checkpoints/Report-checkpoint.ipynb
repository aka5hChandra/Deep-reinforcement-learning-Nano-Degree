<h1>Deep Deterministic Policy Gradient - DDPG </h1>

DDPG is an Actor-Critic method, it is kind of an approximate DQN method for continuous action spaces. Actor-Critic method uses two deep neural networks, called actor and critic. Actor is used to approximate the optimal policy deterministically, i.e to find the best action for a given state, <b><i>argmax<sub>a</sub>Q(s,a)</b></i>.The Critic on the other hand is trained to evaluate optimal action value function by using the actors best believed action.

DDPG uses a replay buffer and soft updates to the target networks.DDPG has two copies of network weights for each network. i.e regular and target networks weights for both actor and critic. Unlike DQN,in DDPG the target networks are updated using a soft updates strategy, slowly blending regular network weights with target network weights. This strategy helps achieve  faster convergence 

<h1> Model architecture </h1>

The Actor network consists of three fully connected layers with batch normalization applied to the input. The network maps states to actions. It uses Leaky-ReLU as activation function for the first two layers and tan for the last layer.

The Critic network consists of four fully connected layers with batch normalization applied at the input. The network maps (state,action) pairs to Q-values. It uses Leaky-ReLU as activation function in the first three layers and no activation function for the last layer.

<h1>Hyperparameters </h1>

BUFFER_SIZE = int(1e6)  # replay buffer size <br/>
BATCH_SIZE = 1024       # minibatch size <br/>
GAMMA = 0.99            # discount factor <br/>
TAU = 1e-3              # for soft update of target parameters <br/>
LR_ACTOR = 1e-4         # learning rate of the actor <br/>
LR_CRITIC = 3e-4        # learning rate of the critic <br/>
WEIGHT_DECAY = 0.0001   # L2 weight decay <br/>

Actor network <br/>
fc1=256 #number of nodes in first hidden layer <br/>
fc2=128,#number of nodes in second hidden layer <br/>
leak=0.01 # leaky ReLUs negtive slope <br/>

Critic network <br/>
fc1=256 #number of nodes in first hidden layer <br/>
fc2=128,#number of nodes in second hidden layer <br/>
fc2=128,#number of nodes in third hidden layer <br/>
leak=0.01 # leaky ReLUs negtive slope <br/>

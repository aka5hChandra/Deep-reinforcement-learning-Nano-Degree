
# Multiple Deep Deterministic Policy Gradient - DDPG

In this project, an environment with two agents playing tennis is solved using Multi Agent DDPG (MADDPG). MADDPG solves multiple Deep Deterministic Policy Gradient(DDPG) algorithms. DDPG is an Actor-Critic method, it is kind of an approximate DQN method for continuous action spaces. Actor-Critic method uses two deep neural networks, called actor and critic. Actor is used to approximate the optimal policy deterministically, i.e to find the best action for a given state, argmaxaQ(s,a)</i>.The Critic on the other hand is trained to evaluate optimal action value function by using the actors best believed action.

DDPG uses a replay buffer and soft updates to the target networks.DDPG has two copies of network weights for each network. i.e regular and target networks weights for both actor and critic. Unlike DQN,in DDPG the target networks are updated using a soft updates strategy, slowly blending regular network weights with target network weights. This strategy helps achieve faster convergence

MADDPG is approached with a cooperative environment, where the agents are concerned about accomplishing a group task and corporate to do so.Each agent’s critic is trained using the observations and actions from both agents, whereas each agent’s actor is trained using just its own observations.



# Model architecture

The Actor network consists of three fully connected layers. The network maps states to actions. It uses ReLU as activation function for the first two layers and tan for the last layer.

The Critic network consists of three fully connected layers. The network maps (state,action) pairs to Q-values. It uses ReLU as activation function in the first three layers and no activation function for the last layer.

state_size = 24 <br />
action_size = 2<br />

Actor networks: <br />
input = 24 # input = state size<br />
fc1 = 64   # number of nodes in first hidden layer <br />
fc2 = 64   # number of nodes in second hidden layer <br />

Critic networks: <br />
input = 52 # input =  (state_size + action_size) x n_agents <br />
fcs1 = 64  # number of nodes in first hidden layer <br />
fc2 = 64   # number of nodes in second hidden layer <br />
fc3 = 64   # number of nodes in third hidden layer <br />
output = 1 <br />

Hyperparameters <br />

GAMMA = 0.99                    # discount factor <br />   
TAU = 5e-2                      # for soft update of target parameters <br />
LR_ACTOR = 5e-4                 # learning rate of the actor <br />
LR_CRITIC = 5e-4                # learning rate of the critic <br />
WEIGHT_DECAY = 0.0              # L2 weight decay <br />
